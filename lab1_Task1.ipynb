```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100000x1 random arrays
exp_data = np.random.exponential(scale=1.0, size=(100000,1))
uniform_data = np.random.rand(100000,1)
normal_data = np.random.randn(100000,1)

# Plot histograms
plt.hist(exp_data, bins=200, density=True, histtype="step", color="green", label="Exponential")
plt.hist(uniform_data, bins=200, density=True, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, bins=200, density=True, histtype="step", color="red", label="Normal")

# Adjust plot view
plt.axis([0, 8, 0, 1])   # Focus on main distribution region
plt.legend(loc="upper right")
plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid of X and Y values
x = np.linspace(-5, 5, 200)
y = np.linspace(-5, 5, 200)
X, Y = np.meshgrid(x, y)

# Compute Z
Z = X**2 + Y**2

# Create 3D plot
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_title("3D Surface: Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Add color bar
fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
df = pd.read_csv('Pokemon.csv', encoding='latin')

# Select relevant features
features = ["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]
data = df[features].dropna()  # remove any missing values

# Calculate Pearson and Spearman correlations
pearson_corr = data.corr(method="pearson")
spearman_corr = data.corr(method="spearman")

# Plot Pearson heatmap
plt.figure(figsize=(8,6))
sns.heatmap(pearson_corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Pearson Correlation (Pokémon Stats)")
plt.show()

# Plot Spearman heatmap
plt.figure(figsize=(8,6))
sns.heatmap(spearman_corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Spearman Correlation (Pokémon Stats)")
plt.show()
```
