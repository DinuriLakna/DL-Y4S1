```python
import pandas as pd

# Load the dataset (adjust the filename if needed)
df = pd.read_csv('Au_nanoparticle_dataset.csv')

# Show the first few rows to check
df.head()
```

```python
# Columns to keep
cols_to_keep = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Create a new filtered dataframe
df_filtered = df[cols_to_keep]

# Show the new dataframe
df_filtered.head()
```

```python
# Display the first 20 rows of the filtered dataframe
df_filtered.head(20)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# Display first 20 samples
df_filtered.head(20)
```

```python
# Calculate mean and standard deviation
stats = df_filtered.agg(['mean', 'std'])

# Calculate quartiles
quartiles = df_filtered.quantile([0.25, 0.5, 0.75])

# Combine results into a single dataframe
summary_stats = pd.concat([stats, quartiles])
summary_stats
```

```python
import matplotlib.pyplot as plt

# Create a 1x4 layout
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# List of features to plot
features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']

# Plot each histogram
for i, col in enumerate(features):
    axes[i].hist(df_filtered[col], bins=20, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the pairplot
sns.pairplot(df_filtered, diag_kind="hist", palette="viridis")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create pairplot with KDE on diagonals
g = sns.pairplot(df_filtered, diag_kind="kde", palette="viridis")

# Overlay histograms on the diagonal
for i, var in enumerate(df_filtered.columns):
    ax = g.diag_axes[i]
    sns.histplot(df_filtered[var], ax=ax, bins=20, color='skyblue', edgecolor='black', alpha=0.5, kde=False)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a base pairplot with KDE diagonals
g = sns.pairplot(df_filtered, diag_kind="kde", palette="viridis")

# Overlay histograms + KDE on the diagonal
for i, var in enumerate(df_filtered.columns):
    ax = g.diag_axes[i]
    sns.histplot(df_filtered[var], ax=ax, bins=20, color='skyblue',
                 edgecolor='black', alpha=0.5, kde=True)

# Add bivariate KDE plots to the lower triangle
g.map_lower(sns.kdeplot, fill=True, cmap="Blues", thresh=0.05, levels=100)

plt.show()
```
